```python
from solver import create_roster_json, Instance, Solver
import pandas as pd
```

```python
#Â Define instance file
demand_baseline = "1.00"
model = "fixed"
week_demand_type = "doublepeak" 
weekend_demand_type = "uniform"
regional_multiplier = 1.5
global_multiplier = 0.8
outsourcing_cost_multiplier = 1.5 

# Define rostering problem
cost_courier=1
min_hours_worked=32
max_hours_worked=48
max_unique_starts = 3
shift_length = 4
working_days = 6
```

# Ouput results to csv

```python
for city in ["berlin", "frankfurt", "lyon", "paris"]:
    for model in ["partflex3", "fixed"]:
        # Create roster json
        create_roster_json(city, demand_baseline, model, week_demand_type, weekend_demand_type, regional_multiplier, global_multiplier, outsourcing_cost_multiplier)

        # Create instance
        instance_name = f'{city}_db={demand_baseline}_weekdt={week_demand_type}_weekenddt={weekend_demand_type}_cout={outsourcing_cost_multiplier}_model={model}_rm={regional_multiplier}_gm={global_multiplier}.json'
        instance = Instance(instance_file=f'rostering_instances/{instance_name}', cost_courier=cost_courier, min_hours_worked=min_hours_worked, max_hours_worked=max_hours_worked, max_unique_starts=max_unique_starts, shift_length=shift_length, working_days=working_days)
        
        # Solve instance
        solver = Solver(instance)
        base_results = solver.solve_base()
        roster_results = solver.solve_roster()

        # Save results in CSV files
        base_results_df = pd.DataFrame(base_results, index=[0])
        roster_results_df = pd.DataFrame(roster_results, index=[0])
        results_df = pd.concat([base_results_df, roster_results_df], axis=0)
        results_df.to_csv(f'../rostering_results/{instance_name}_jonny.csv', index=False)
        
```
